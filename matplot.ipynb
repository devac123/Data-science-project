# matplotlib

```python
import matplotlib.pyplot as pl
```

```python
a = list(range(1,5))
```

```python
a
```

```python
d = [2,3,4,5]
```

```python
d
```

```python
pl.plot(a,d, color = 'c',marker='o',markersize='15',linestyle ='--',linewidth ='3')
pl.title('learning matplotlib',fontsize =20)
pl.xlabel('x label',fontsize =20)
pl.ylabel('do the working',fontsize =20)
pl.legend(['hello'],loc = 4)
```

```python
from matplotlib import style
```

```python
style.use('ggplot')
pl.plot(a,d,label ="hello", color = 'c',marker='o',markersize='15',linestyle ='--',linewidth ='3')
pl.title('learning matplotlib',fontsize =20)
pl.xlabel('x label',fontsize =20)
pl.ylabel('do the working',fontsize =20)
pl.legend(loc = 3)
```

```python
f = [5,4,3,4]
style.use('ggplot')
pl.plot(a,d,label ="hello", color = 'c',marker='o',markersize='15',linestyle ='--',linewidth ='3')
pl.plot(a,f,label ="bye", color = 'g',marker='o',markersize='15',linestyle =':',linewidth ='3')
pl.title('learning matplotlib',fontsize =20)
pl.xlabel('x label',fontsize =20)
pl.ylabel('do the working',fontsize =20)
pl.legend(loc = 3)
pl.grid(color='k', linestyle='-',linewidth=3)
```
